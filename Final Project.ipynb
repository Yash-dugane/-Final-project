```python
You have to design a Food Ordering app for a restaurant.



The application will have a log-in for admin and users.



Admin will have the following functionalities:

Add new food items. Food Item will have the following details:
FoodID //It should be generated automatically by the application.
Name
Quantity. For eg, 100ml, 250gm, 4pieces etc
Price
Discount
Stock. Amount left in stock in the restaurant.
Edit food items using FoodID.
View the list of all food items.
Remove a food item from the menu using FoodID.




The user will have the following functionalities:

Register on the application. Following to be entered for registration:
Full Name
Phone Number
Email
Address
Password
Log in to the application
The user will see 3 options:
Place New Order
Order History
Update Profile
Place New Order: The user can place a new order at the restaurant.
Show list of food. The list item should as follows:
1. Tandoori Chicken (4 pieces) [INR 240]
2. Vegan Burger (1 Piece) [INR 320]
3. Truffle Cake (500gm) [INR 900]
Users should be able to select food by entering an array of numbers. For example, if the user wants to order Vegan Burger and Truffle Cake they should enter [2, 3]
Once the items are selected user should see the list of all the items selected. The user will also get an option to place an order.
Order History should show a list of all the previous orders
Update Profile: the user should be able to update their profile.
```

```python
import random
class Foodie_Zone:
    def __init__(self):
        self.Admin_Login = False
        self.Admin_Details = {}
        #self.Food=[]
        self.Foods = {}
        self.User_Details = {}
        self.Order_History =[]
        self.User_Login = False
        
    def Admin_Registration(self):
        self.Admin_Name = input("\n Enter Your Name:- ")
        a = int(input("\n Enter Your Mobile Number:- "))
        if len(str(a))==10:
            self.Contact_Number = a
        else:
            print("\n Please Enter A Valid Number:- ")
        
        self.Admin_Mail_ID = input("\n Enter Your Mail-ID:- ")
        self.Admin_Password = input("\n Set Your Password:- ")
        s = {}
        s["Admin_Name"] = self.Admin_Name
        s["Contact Number"] = self.Contact_Number
        s["Admin Mail_ID"] = self.Admin_Mail_ID
        s["Admin Password"] = self.Admin_Password
        self.Admin_Details[self.Contact_Number] = s
        print("\n Registration Completed")
        
        
    def Login_Admin(self):
        num = int(input("\n Enter Your Mobile Number:- "))
        if num in self.Admin_Details:
            d = input("\n Enter The Password:- ")
            if self.Admin_Details[num]["Admin Password"]== d:
                self.Admin_Login = True
                print("\n Login Sucessfull")
                
        else:
            print("\n No Match Found,Please Enter Valid Details")
                
    def Add_New_Food(self):
        self.Food= []
        self.Food_ID = random.randrange(1,100,1)
        f = input("\n Food Name:-")
        self.Food.append(f)
        g = input("\n Quantity[into peices,grams,ml]:-")
        self.Food.append(g)
        h = input("\n Price:-")
        self.Food.append(h)
        self.Foods[self.Food_ID] = self.Food
        print("\n New Food Item Added Sucessfully")
        print(self.Foods)
        
    def Discount(self):
        j = input(" Hurry-up,and get the discount of 10% on every order\n To grab the Discount enter 10% off:-")
        if j == "10% off":
            print("\n Discount Applied")
            
        else:
            print("\n You have'nt applied for any offer")
            
            

            
    def Stock(self):
        print(" \n  Total no.of food items :- ",len(list(self.Foods.keys())))
        k=1
        for i in self.Foods:
            print("\n The following list show's available items  ")
            print("\n  S.no-",k,"\n")                 
            print("  Food ID        :- ",i)
            print("  Food Name      :- ",self.Foods[i][0])
            print("  Food Quantity  :- ",self.Foods[i][1])
            print("  Food Price     :- ",self.Foods[i][2])
      
            k+=1
        
    def Edit_Foods(self):
        l = int(input("\n Enter Food ID:- "))
        if l in self.Foods:
            z=int(input("\n 1 To Change Food ID \n 2 To Change Food Name  \n 3 To Change Food Quantity \n 4 To Change Food Price"))
            if z==1:
                b=self.Foods[l]
                self.Foods.pop(l)
                c=input(" \n  Enter new Food ID:- ")
                self.Foods[c]=b
                print(" \n  Food ID changed sucessfully ")
            if z==2:
                self.Foods[l][0]=input("  Enter Food Name:- ")
                print(" \n  Food Name Changed Sucessfully ")
            if z==3:
                self.Foods[l][1]=int(input("  Enter Food Quantity:- "))
                print("\n  Food Quantity Changed Sucessfully ")
            if z==4:
                self.Foods[l][2]=int(input("  Enter Food Price:- "))
                print("\n Food Price Changed Sucessfully ")
                
        else:
            print("\n  Invalid Food ID ")
            
        
    def List_of_Food_Items(self):
        print(" Food List is as Follow \n ",self.Foods.values())
        

    def Remove_Food(self):
        x=int(input(" \n Enter Food ID:- "))
        if x in self.Foods:
            p=input(" \n  To Delete this Food ID enter 'YES':- ")
            if p=="YES":
                self.Foods.pop(x)
                print(" \n Food Deleted Sucessfully ")
            if p=="NO":
                    pass
        else:
            
            print(" \n Food ID not found ")
            
            
    def User_Registration(self):
        
        self.User_Name= input("\n Enter Name:- ")
        v=int(input("\n Enter mobile number:- "))
        if len(str(v))==10:
            self.User_Number=v
        else:
            print("\n Invalid number")
        
        self.User_Mail= input("\n Enter Mail ID:- ")
        self.User_Pass= input("\n Set Password:- ")
        self.User_Adress= input("\n Enter Adress:- ")
        q={}
        q["User_Name"]= self.User_Name
        q["User_Number"]= self.User_Number
        q["User_Mail"]= self.User_Mail
        q["User_Pass"]= self.User_Pass
        q["User_Adress"]= self.User_Adress
        self.User_Details[self.User_Number]=q
        print("\n Registration completed sucessfully ")
        
        
    def Login_User(self):
        num= int(input("\n Enter mobile number:- "))
        if num in self.User_Details:
            w= input("\n Enter Password:- ")
            if self.User_Details[num]["User_Pass"]== w:
                self.User_Login= True
                print("\n Login Sucessful")
                w= int(input("\n 1 Place New Order:-  \n 2 Order History:-  \n 3 Update Profile:- \n\n Enter sr_no:- "))
                if w==1:
                    print("\n Here is list of all food items \n " ,list(self.Foods.values()))
                    print("\n Enter Food ID to buy Food \n ", self.Foods)
                    t= int(input("\n Enter Food ID:- "))
                    if t in self.Foods:
                        r= input("\n Enter 'YES' to confirm:-")
                        if r=="YES":
                            g=self.Foods.pop(t)
                            
                            self.Order_History.append(g)
                            print("\n Order placed succesfully ")
                        else:
                            print("\n Sorry for your inconveniance")
                    else:
                        print("\n Food ID not found ")
                        
                if w==2:
                    print("\n Order History:- ", self.Order_History)
                    
                if w==3:
                    y= int(input("\n Enter mobile number:- "))
                    if y in self.User_Details:
                        e= list(self.User_Details.values())[0]
                        n= int(input(" \n 1 Change Name:- \n 2 Change Mail ID:- \n 3 Change Number:- \n 4 Change Password:- \n 5 Change Address:- \n\n Enter no:-  "))
                        if n==1:
                            e["user_name"]= input("\n Enter new user name:- ")
                            print("\n Profile updated sucessfully ", self.User_Details)
                    
                        elif n==2:
                            e["user_name"]= input("\n Enter new mail:- ")
                            print("\n Profile updated sucessfully ", self.User_Details)
                    
                        elif n==3:
                            e["user_number"]=int(input("\n Enter new number:- "))
                            print("\n Profile updated sucessfully ", self.User_Details)
                     
                        elif n==4:
                            e["user_pass"]= input("\n Enter new password:- ")
                            print("\n Profile updated sucessfully ", self.User_Details)
                     
                        elif n==5:
                            e["user_address"]= input("\n Enter new address:- ")
                            print("\n Profile updated sucessfully ", self.User_Details)
                            
                    else:
                        print("Entered wrong number")
                     
                            
                      
                    
        else:
            print("\n User Not Found")
            
            
            
    def menu(self):
        i= True
        while i:
            u= int(input("\n 1 Admin Registration:- \n 2 Admin Login:- \n 3 Add New Food Item:- \n 4 Know the Discount:- \n 5 Current stock:-  \n 6 Edit Food Items:- \n 7 List of Food Items:- \n 8 Remove Food:- \n 9 User Registration:- \n 10 User Login:- \n\n Enter sr_no:- "))
            if u==1:
                self.Admin_Registration()
            if u==2:
                self.Login_Admin()
            if u==3:
                self.Add_New_Food()
            if u==4:
                self.Discount()
            if u==5:
                self.Stock()
            if u==6:
                self.Edit_Foods()
            if u==7:
                self.List_of_Food_Items()
            if u==8:
                self.Remove_Food()
            if u==9:
                self.User_Registration()
            if u==10:
                self.Login_User()
                
        else:
            print(" Invalid Input ")
            
            
            
            
            
```
